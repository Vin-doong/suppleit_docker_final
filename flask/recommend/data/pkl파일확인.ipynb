### TF분류된 단어들의 점수와 IDF값 

```python
import pickle

with open("tfidf_vectorizer_비타민.pkl", "rb") as f:
    data = pickle.load(f)

# 단어와 IDF 값을 매칭하여 출력
idf_dict = {word: data.idf_[index] for word, index in data.vocabulary_.items()}

# IDF 값이 낮은 순으로 출력 (자주 등장하는 단어일수록 중요도가 높다고 판단)
sorted_idf = sorted(idf_dict.items(), key=lambda x: x[1])

for word, idf in sorted_idf:
    print(f"단어: {word}, IDF 값: {idf}")
```
